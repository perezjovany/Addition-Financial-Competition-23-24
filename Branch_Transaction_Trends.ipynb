```python
import pandas as pd

# Load Branch Level Data and Member Level Data
branch_data = pd.read_csv("data/branch_data.csv")
member_data = pd.read_csv("data/member_data.csv")

# Merge the datasets based on "BranchCategory" column
merged_data = pd.merge(branch_data, member_data, on="BranchCategory")
```

```python
#Create column for total transcations of a row
branch_data['Total_Transactions'] = branch_data[['ATM', 'Bill Payment', 'Cash', 'Draft', 'ACH', 'Fee', 'Credit/Debit Card', 'Home Banking', 'Dividend']].sum(axis=1)

# Convert 'EOM TRANS DATE' to a datetime object for date-based operations
branch_data['EOM_TRANS_DATE'] = pd.to_datetime(branch_data['EOM_TRANS_DATE'])

# Sort the data by date
branch_data = branch_data.sort_values(by='EOM_TRANS_DATE')

# Calculate monthly transaction counts for each branch category
monthly_demand = branch_data.groupby(['BranchCategory', branch_data['EOM_TRANS_DATE'].dt.to_period('M')]).sum()

# Reset the index to make 'BranchCategory' and 'EOM TRANS DATE' as columns
monthly_demand = monthly_demand.reset_index()

# Pivot the data to have 'BranchCategory' as columns
monthly_pivot = monthly_demand.pivot(index='EOM_TRANS_DATE', columns='BranchCategory', values='Total_Transactions')

# Calculate the percentage change in demand over time for each branch category
demand_change = monthly_pivot.pct_change()

# Calculate the mean percentage change for each branch category
mean_demand_change = demand_change.mean()

# Sort the branch categories by their mean percentage change in descending order
sorted_branch_categories = mean_demand_change.sort_values(ascending=False)

# Print the branch categories with the highest increase in demand
print("Branch Categories with the Highest Increase in Demand:")
print(sorted_branch_categories.head())
```
